```python
import scrapy
from scrapy.crawler import CrawlerProcess
```

```python
page_size = 10
count = 32
make_url = lambda offset: f"https://www.booking.com/searchresults.ru.html?label=gen173nr-1FCAEoggJCAlhYSDNYBGjCAYgBAZgBIbgBB8gBD9gBAegBAfgBC5ICAXmoAgM&lang=ru&sid=60ecc339457c36fb3581fb59e5fdbe53&sb=1&src=index&src_elem=sb&error_url=https%3A%2F%2Fwww.booking.com%2Findex.ru.html%3Flabel%3Dgen173nr-1FCAEoggJCAlhYSDNYBGjCAYgBAZgBIbgBB8gBD9gBAegBAfgBC5ICAXmoAgM%3Bsid%3D60ecc339457c36fb3581fb59e5fdbe53%3Bsb_price_type%3Dtotal%26%3B&ss=%D0%B2%D0%BE%D1%80%D0%BE%D0%BD%D0%B5%D0%B6&checkin_monthday=&checkin_month=&checkin_year=&checkout_monthday=&checkout_month=&checkout_year=&group_adults=2&group_children=0&no_rooms=1&from_sf=1&ss_raw=dj&dest_id=&dest_type=&search_pageview_id=fafc3ab5333e009f&search_selected=false&rows=20&offset={offset}"
page_urls = [make_url(p * page_size) for p in range((count // page_size) + 1)]
```

```python
class BookingSpider(scrapy.Spider):
    name = "booking"
    start_urls = page_urls

    def parse(self, response):
        for hotel in response.css('.sr_item'):
            yield{
                'title': hotel.css('.sr-hotel__name::text').extract_first(),
                'type': hotel.css('.sr-hotel__type::text').extract_first(),
                'href': hotel.css('.hotel_name_link.url::attr(href)').extract_first(),
            }
```

```python
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.11; rv:57.0) Gecko/20100101 Firefox/57.0'
})

process.crawl(BookingSpider)
process.start()
```
