```python
sber = pd.read_csv('./sberbank-utf.csv')
```

```python
s = sber['name']
wage = sber[sber['name'] == 'Средняя зарплата'].sort_values('date')
ax = None
regions = list(wage['region'].unique())
regions.append('Россия')
for region in regions:
    is_pivot = (region == 'Россия')
    ax = wage[wage['region'] == region].plot(
        x='date', 
        y='value', 
        ax=ax, 
        legend=False, 
        linewidth=5 if is_pivot else 1,
        figsize=(20,20))
```
