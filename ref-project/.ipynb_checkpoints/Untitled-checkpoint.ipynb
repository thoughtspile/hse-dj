```python
import pandas as pd
import json
%matplotlib inline
```

```python
cities = pd.read_json('./city-list.json')
cities = cities [2:]
cities['pop'] = pd.to_numeric(cities['pop'])
# cities['founded_y'] = pd.to_numeric(cities['founded_y'])
cities = cities.sort_values('pop', ascending=False)
big_cities = cities[cities['pop'] > 300000]
big_cities.to_json('./city-list-big-clean.json', orient='records', force_ascii=False)
big_cities
```

```python
d = json.load(open('./culture.json', 'r'))
```

```python
data = [e['data']['general'] for e in d]
rm = ['documents', 'conNumber', 'borderInfo']
for row in data:
    for r in rm:
        if r in row:
            del row[r]
data[0]
cult = pd.io.json.json_normalize(data)
```

```python
c = cult['address.mapPosition.type']
# w_location = cult[~cult['address.mapPosition.coordinates'].isnull() & ~cult['additionalCoordinates'].isnull()]
w_location = cult[~cult['address.fullAddress'].isnull()]
del w_location['address.mapPosition.type']
del w_location['regNumber']
del w_location['regimeInfo']
del w_location['photo.title']
del w_location['photo.preview']
del w_location['id']
del w_location['isActual']

w_city = w_location['address.fullAddress'].str.match('.*(г|город)[\.\s]\s*(\w+)') | False
spb = w_location['region.value'].str.match('.*Санкт')
mow = w_location['region.value'].str.match('.*Москва')

w_location['city'] = w_location['address.fullAddress'].str[0:0]

w_location.loc[w_city, 'city'] = w_location['address.fullAddress'].str.extract('.*(г|город)[\.\s]\s*([а-яёЁА-Я\-]+)', expand=False)[1].str.strip()
w_location.loc[spb, 'city'] = 'Санкт-Петербург'
w_location.loc[mow, 'city'] = 'Москва'
```

```python
w_location = w_location[w_location['city'].str.len() > 0]
w_location.groupby('city').count().sort_values('name', ascending=False)
```

```python
c_count = w_location.groupby('city').count().sort_values('name', ascending=False).reset_index()[['city', 'address.fullAddress']]
def get_pop(c):
    p = cities[cities['name'] == c]['pop']
    return 0 if p.empty else p.iat[0]
c_count['pop'] = c_count['city'].map(get_pop)
c_count
```

```python
c_count = c_count[c_count['pop'] > 0]
c_count['perc'] = c_count['address.fullAddress'] / c_count['pop']
# c_count = c_count[c_count['pop'] > 30000]
c_count.sort_values('perc', ascending=False)
```

```python
c_count.sort_values('perc', ascending=False)[50:100]
```

```python
weather = pd.read_json('./weather.json')
data_cols = ['month_n', 'year', 't_avg', 'rain_mm']
for c in data_cols:
    weather[c] = pd.to_numeric(weather[c])
# все значения не месте
print(weather.isnull().sum())
# считаем среднемесячные температуры
weather_m = weather.groupby(['city_name', 'month_n']).mean().reset_index()
```

```python
# отсортируем по температуре – похоже на правду
by_temp = weather.groupby('city_name').mean().sort_values('t_avg', ascending=False)
by_temp[['t_avg', 'rain_mm']]
```

```python
# кое-кто убежал, но это Сочи, ему можно
weather.groupby('city_name').mean()[['rain_mm', 't_avg']].plot.scatter(x='rain_mm', y='t_avg')
```

```python
# идея визуализации - сравнивать города по погоде. 
def city_weather_vs(c1, c2):
    ax = weather[weather['city_name'] == c2][['rain_mm', 't_avg']].plot.scatter(x='rain_mm', y='t_avg', color='grey', edgecolors='none')
    weather[weather['city_name'] == c1][['rain_mm', 't_avg']].plot.scatter(x='rain_mm', y='t_avg', color='red', ax=ax, alpha=.2, edgecolors='none')
# да, я бы с удовольстивем оказался в Сочи
city_weather_vs('Сочи', 'Москва')
# разрушаем стереотипы - погода в Питере и Москве одинаковая
city_weather_vs('Санкт-Петербург', 'Москва')
```

```python
uni = pd.read_json('./uni.json')
# нормальные даные, даже 568 проходных баллов есть
print(uni.count())
data_cols = ['exam', 'price', 'qty']
for c in data_cols:
    uni[c] = pd.to_numeric(uni[c])
```

```python
b_city = uni[~uni['qty'].isnull()].groupby('city').sum().sort_values('qty', ascending=False)
b_city
```

```python
uni_count = b_city.reset_index()
def get_pop(c):
    p = cities[cities['name'] == c]['pop']
    return 0 if p.empty else p.iat[0]
uni_count['pop'] = uni_count['city'].map(get_pop)
uni_count['per_cap'] = uni_count['qty'] / uni_count['pop']
uni_count[uni_count['pop'] > 0].sort_values('per_cap', ascending=False)
# Томск и Иркутск ок, Белгород вроде такой динамичный, но Махачкала?!
```
