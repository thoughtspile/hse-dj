```python
import pandas
import json
from pandas.io.json import json_normalize
%matplotlib inline
```

```python
attr = pandas.read_excel('voronezh/attractions.xlsx', header=1)
# сразу удалим лишние столбцы
del attr['Unnamed: 6']
del attr['№ п/п']
del attr['Документ о принятии на гос. охрану']
del attr['Категория охраны']
attr
```

```python
# так бырираем строки
attr[1:10]
# так выбираем один столбец
attr['Наименование памятника']
# так выбираем несколько столбцов
attr[['Наименование памятника', 'Датировка']]
# всё вместе
attr[['Наименование памятника', 'Датировка']][0:5]
```

```python
# из столбцов можно делать булевы векторы 
has_20_y = attr['Датировка'].str.match(r'^19\d\d')
has_20_c = attr['Датировка'].str.match(r'.*XX')
# для таких векторов перегружены поитовые операторы
is_20th = has_20_y | has_20_c
marx_st = attr['Адрес'].str.match('.*Карла Маркса')
# доступ к фрейму по булеву вектору достаёт только те строки, на месте которых True
attr[is_20th & marx_st]
```

```python
# так не сработает, потому что несвязные данные завернуты в объект и пандас не может определить схему
# pandas.read_json('voronezh/booking.json')
# поэтому открываем json рукми, потом делаем фрейм из нужного поля
booking = pandas.DataFrame(json.load(open('voronezh/booking.json', 'r'))['b_hotels'])
```

```python
# с первого раза гистограмма не построилась, потому что b_review_score - строка. Сделаем число:
booking['b_review_score'] = pandas.to_numeric(booking['b_review_score'])
booking['b_u_total_price_rounded'].hist()
```

```python
# похоже, тут корреляция!
booking.plot.scatter(x='b_review_score', y='b_u_total_price_rounded')
# хотя слабенькая
booking[['b_review_score', 'b_u_total_price_rounded']].corr()
```

```python
# теперь объединим три файла, данные в которых пересекаются
def load_fs(name):
    # на этот раз данные спрятаны поглубже
    j = json.load(open('voronezh/' + name + '.json', 'r'))['response']['group']['results']
    # в джисоне вложенная структура, а пандас может только плоскую. 
    # json_normalize разворачивает объект
    df = json_normalize([r for r in j if r['displayType'] == 'venue'])
    # объединять будем по уникальному id кафе
    df.set_index('id', inplace=True)
    return df
# грузим и сцепляем
fs1 = load_fs('fs1')
fs2 = load_fs('fs2')
fs3 = load_fs('fs3')
fs = fs1.append(fs2).append(fs3)
fs.head()
```

```python
# теперь испртим данные
fs.loc['5a8146980a464d412808b7c1', 'venue.location.lng'] = 10
# на графике сразу виден подвох
fs.plot.scatter(x='venue.location.lng', y='venue.location.lat')
```

```python
# выкинем лишние данные
fs = fs[fs['venue.location.lng'] > 25]
# и насладимся кофейной картой Воронежа
fs.plot.scatter(x='venue.location.lng', y='venue.location.lat')
```

```python
# теперь сложим все данные в один файл
fs.to_json('res.json', orient='records')
```
